## Adjust CFPB zip code data because for some they truncate the last two digits. Impute these at random.

```python
cfpbRows.iloc[2].zip_code[3:5]
```

```python
import random
def imputeTruncatedZipCodes(x):
    #print x
    if str(x)=='None':
        #print "Problem: ", x
        return x
    if x[-2:]=='XX':
        try:
            prefix = int(x[0:3])
            #print "Prefix: ", prefix
            matches = filter(lambda y: int(y/100)==prefix, zipCountyRows.zip_code)
            return int(matches[int(random.random()*len(matches))])
        except:
            return None ### This is damaged data. Prepare to discard.
    else:
        try:
            return int(x)
        except:
            return None #Still problematic

print imputeTruncatedZipCodes('85208')
print imputeTruncatedZipCodes('3-007')
```

```python
import multiprocessing
p = multiprocessing.Pool(6)
#cfpbRows.zip_code = map(imputeTruncatedZipCodes, cfpbRows.zip_code)
cfpbRows.zip_code = p.map(imputeTruncatedZipCodes, cfpbRows.zip_code)
```

### Translate cfpb zip codes to county names

```python
import random
def translateZipToCountyCode(x):
    try:
        x=int(x)
        codes = zipCountyRows.trunc_county_code[x]
        try:
            return codes.iloc[int(random.random()*len(codes))] #Return one at random if there is more than one
        except AttributeError: #If there's just one, return that
            return codes
    except KeyError:
        return None
```

```python
countyIncomeRows = pd.DataFrame(cur.fetchall(), columns = ["CountyName", "SumAGI", "SumNumReturns"])
countyIncomeRows.head()
```

```python
from __future__ import division
countyIncomeRows['AverageAGI']=[x/y for (x, y) in zip(countyIncomeRows.SumAGI, countyIncomeRows.SumNumReturns)] #NOTE: These are measured in THOUSANDS
```

```python
def getAverageAGI(x):
    if x=="LaSalle Parish":
        x="La Salle Parish"
    elif x=="Petersburg Borough":
        x="Petersburg Census Area"
    try:
        return countyIncomeRows.AverageAGI[x]
    except:
        y = x+" County"
        return countyIncomeRows.AverageAGI[y]
            

cfpbRows['CountyAGI'] = map(getAverageAGI, cfpbRows.county_name)
```

```python
#Fix the Tags column to be more convenient

def fixTags(x):
    if str(x)=='None':
        return "Neither"
    elif x=="Older American":
        return "OlderAmerican"
    elif x=="Servicemember":
        return "Servicemember"
    elif ("Servicemember" in x) and ("Older American" in x):
        return "Both"

    
cfpbRows.tags = map(fixTags, cfpbRows.tags)
cfpbRows.head()
```
